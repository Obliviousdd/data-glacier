# Data Wrangling

```python
import pandas as pd
import numpy as np
city = pd.read_csv('City.csv')
city.head()
```

```python
cab = pd.read_csv('Cab_Data.csv')
cab.head()
```

```python
df = pd.merge(city,cab)
df
```

```python
customer = pd.read_csv('Customer_ID.csv')
customer.head()
```

```python
trans = pd.read_csv('Transaction_ID.csv')
trans.head()
```

```python
df2 = pd.merge(customer,trans)
df2
```

```python
df3 = pd.merge(df,df2)
df3
```

```python
#check dupicates
duplicate = df3[df3.duplicated()]
duplicate
```

```python
#check NA
df3 = df3.dropna()
df3
```

```python
df3['profit'] = df3['Price Charged'] - df3['Cost of Trip']
df3.memory_usage().sum()/2**20
```

```python
# calculate interquartile range
q25, q75 = np.percentile(df3['profit'], 25), np.percentile(df3['profit'], 75)
iqr = q75 - q25
iqr
```

```python
# calculate the outlier cutoff
cut_off = iqr * 1.5
lower, upper = q25 - cut_off, q75 + cut_off
lower,upper
```

```python
# identify outliers
outliers = [x for x in df3['profit'] if x < lower or x > upper]
# remove outliers
outliers_removed = [x for x in df3['profit'] if x >= lower and x <= upper]
```

```python
df3['male'] = df3.Gender.map({'Male':1, 'Female':0})
df3['company']= df3.Company.map({'Pink Cab':1,'Yellow Cab':0})
df3['pay'] = df3.Payment_Mode.map({'Cash':1,'Card':0})
```

```python
df3.drop(['Company','Gender','Payment_Mode'],axis = 1,inplace = True)
```

```python
df3.set_index('Transaction ID', inplace = True)
df3
```

```python
df3.columns = ['city','population','users','date','km','price','cost','customer_id','age','income','margin','male','company','pay']
df3
```

```python
pink  = df3[df3['company']==1]
user1 = pink[['date','customer_id']].groupby('date').count()
margin1 = pink[['date','margin']].groupby('date').sum()
d1 = pd.merge(user1,margin1, left_index = True, right_index = True)
d1
```

```python
yellow  = df3[df3['company']==0]
user2 = yellow[['date','customer_id']].groupby('date').count()
margin2 = yellow[['date','margin']].groupby('date').sum()
d2 = pd.merge(user2,margin2, left_index = True, right_index = True)
d2
```

```python
user = pd.merge(d1,d2,left_index = True, right_index = True)
user.columns = ['pink_users','margin_p','yellow_users','margin_y']
user['pink_prop'] = user['pink_users'].pct_change()
user['margin_p_prop'] = user['margin_p'].pct_change()
user['yellow_prop'] = user['yellow_users'].pct_change()
user['margin_y_prop'] = user['margin_y'].pct_change()
user
```
